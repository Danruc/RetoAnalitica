En este trabajo se revisarán y analizar los datos de los tweets publicados, los cuales hacen referencia a la pandemia del Covid-19.

Primeramente, se utilizarán las librerías de pandas, matplotlib y seaborn para poder manipular los datos de una manera mas sencilla. Pandas se usara para leer el documento *”covid19_tweets.csv”*, además de hacer análisis sencillos como media, mediana y moda o correlación de datos. 

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np; np.random.seed(0)
import seaborn as sns
```

Con la función `.read_csv` se puede leer el archivo csv y guardarlo como una dataframe en la variable `df`. Posteriormente, se hizo una segunda dataframe únicamente con las columnas *user_followers*, *user_friends* y *user_favourites*, puesto que son las que tienen valores numéricos. 

```python
df = pd.read_csv("covid19_tweets.csv")
newdf = df.copy()
newdf = newdf.drop(['user_name','user_location','user_description','user_created','user_verified','date','text','hashtags','source','is_retweet'], axis=1)
```

Después de haber leído y guardado los valores del archivo csv, usaremos la librería **pandas** y **numpy** para obtener información básica acerca de los datos. El método `.describe()` sirve para obtener la media, mediana, moda, desviación estándar, etc. Por otra parte, la misma función de la libreria **numpy** se usó para analizar valores no numéricos como *user_name* o *hashtags*.

```python
#pandas
print(df.describe())
#numpy
df.describe(include=np.object).transpose()
```

La librería **pandas** contiene un método `.corr()` para poder evaluar la correlación de los datos. Este puede ser modificado dependiendo el estilo con el que se puede encontrar la correlación. En esta ocasión se realizó la correlación con *Pearson*.

```python
print(df.corr(method='pearson'))
```

Esta misma correlación se hizo con la librería **seaborn**, solo que esta vez con el método `.pairplot()` arroja el resultado de una manera grafica y visual, en lugar de ser numérica. 

```python
sns.pairplot(df)
plt.show()
```

Ahora se realizarán tres *whiskers plot* con ayuda de la librería **matplotlib** . Se graficarán los tres *boxplots* en una misma figura. Esto se hará con el metodo `boxplot()`.

```python
a = df['user_followers']
b = df['user_friends']
c = df['user_favourites']
bplot = [a, b, c]
fig = plt.figure(figsize=(10,7))
ax = fig.add_axes([0, 0, 1, 1])
bp = ax.boxplot(bplot)
plt.show()
```

Con ayuda de la librería **numpy** se harán tres histogramas para las columnas con valores numéricos. Estas columnas son: 
- user_followers
- user_friends
- user_favourites 

El método que se ocupa en estos tres casos es `plt.hist()`.

```python
np.random.seed(10**7) 
mu = 121 
sigma = 21
num_bins = 100
   
n, bins, patches = plt.hist(df["user_followers"], num_bins, density = 1, color ='green', alpha = 0.7) 
   
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2)) 
  
plt.plot(bins, y, '--', color ='black') 
  
plt.xlabel('X-Axis') 
plt.ylabel('Y-Axis') 
  
plt.title('matplotlib.pyplot.hist() function Example\n\n', 
          fontweight ="bold") 
  
plt.show() 
```

```python
np.random.seed(10**7) 
mu = 121 
sigma = 21
num_bins = 100
   
n, bins, patches = plt.hist(df["user_friends"], num_bins, density = 1, color ='green', alpha = 0.7) 
   
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2)) 
  
plt.plot(bins, y, '--', color ='black') 
  
plt.xlabel('X-Axis') 
plt.ylabel('Y-Axis') 
  
plt.title('matplotlib.pyplot.hist() function Example\n\n', 
          fontweight ="bold") 
  
plt.show() 
```

```python
np.random.seed(10**7) 
mu = 121 
sigma = 21
num_bins = 100
   
n, bins, patches = plt.hist(df["user_favourites"], num_bins, density = 1, color ='green', alpha = 0.7) 
   
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2)) 
  
plt.plot(bins, y, '--', color ='black') 
  
plt.xlabel('X-Axis') 
plt.ylabel('Y-Axis') 
  
plt.title('matplotlib.pyplot.hist() function Example\n\n', 
          fontweight ="bold") 
  
plt.show() 
```

Finalmente, se hará un mapa de calor con la librería **seaborn** para poder visualizar la correlación de los datos. El metodo que se ocupa es `sns.heatmap()`. Aquí se ocupa el dataframe *newdf*, debido a que las columnas con valores no numéricos tuvieron que ser eliminadas para que el método funcionara. 

```python
grid_kws = {"height_ratios": (.9, .05), "hspace": .3} 
f, (ax, cbar_ax) = plt.subplots(2, gridspec_kw=grid_kws)
ax = sns.heatmap(newdf, cmap="YlGnBu", ax=ax, cbar_ax=cbar_ax, cbar_kws={"orientation": "horizontal"})
plt.show()
```
